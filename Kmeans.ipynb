```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

pd.set_option('display.max_columns', 30)

dados = load_breast_cancer()

x = pd.DataFrame(dados.data, columns=[dados.feature_names])
y = pd.Series(dados.target)

```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
# normalizando
normalizador = MinMaxScaler(feature_range=(0, 1))
x_norm = normalizador.fit_transform(x)
```

```python
from sklearn.cluster import KMeans

# Criação do modelo:
modelo = KMeans(n_clusters=2, random_state=16)
modelo.fit(x_norm)

# Imprimindo resultado:
print(modelo.cluster_centers_) # centroid
```

```python
print(modelo.predict(x_norm))
```

```python
rotulos_kmeans = modelo.labels_
rotulos_kmeans
```

```python
import numpy as np
acertos = np.sum(rotulos_kmeans == y)
print(f'ACERTOS: {acertos}')
print(f'TOTAL DE DADOS: {y.count()}')

acuracia = (acertos * 100) / y.count()
print(f'TAXA DE ACERTOS EM RELAÇÃO AO TOTAL: {acuracia}')
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y, rotulos_kmeans ))
```
